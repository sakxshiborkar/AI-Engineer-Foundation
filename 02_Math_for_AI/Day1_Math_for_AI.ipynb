### ✨ Concept Summary
In AI and Machine Learning, we represent almost everything (images, words, signals) as **vectors** — numerical lists that describe features.  
- **Scalar:** A single number (e.g., temperature, speed, brightness).  
- **Vector:** A list of numbers (e.g., [3, 4] → position in 2D space).  
- **Norm:** The "length" of a vector (distance from origin).  
- **Dot Product:** Measures alignment or similarity between two vectors.


### 🔢 Key Formulas

‖v‖ = √(x² + y² + z²)  
**Dot Product:**  
a · b = |a||b|cosθ = Σ (aᵢ * bᵢ)


```python
# 🧮 Vector Operations with NumPy

import numpy as np

# Define two vectors
a = np.array([2, 3])
b = np.array([4, 5])

# Calculate magnitude (norm)
magnitude_a = np.linalg.norm(a)
magnitude_b = np.linalg.norm(b)

# Dot product
dot_product = np.dot(a, b)

print("Vector a:", a)
print("Vector b:", b)
print("‖a‖ =", round(magnitude_a, 2))
print("‖b‖ =", round(magnitude_b, 2))
print("Dot Product:", dot_product)
```

### 🧩 Observation
- If dot product > 0 → Vectors point in roughly the same direction.  
- If dot product < 0 → Opposite directions.  
- If dot product = 0 → They are perpendicular (no alignment).

## 🧩 Practice Section — Vector Mastery
Let's apply what you've learned by solving small, practical problems.

We'll cover:
1. **Vector Normalization** — making vectors unit length  
2. **Angle Between Two Vectors** — using dot product formula  
3. **Visualizing Vectors in 2D** — understanding direction visually


```python
# 🧮 1. Vector Normalization (Unit Vector)
import numpy as np

v = np.array([3, 4])
norm_v = np.linalg.norm(v)
unit_v = v / norm_v

print("Original Vector:", v)
print("Magnitude:", norm_v)
print("Normalized (Unit) Vector:", unit_v)
```

```python
# 📐 2. Angle Between Two Vectors
a = np.array([2, 3])
b = np.array([4, 5])

dot = np.dot(a, b)
norm_a = np.linalg.norm(a)
norm_b = np.linalg.norm(b)

cos_theta = dot / (norm_a * norm_b)
angle_deg = np.degrees(np.arccos(cos_theta))

print("Angle between vectors (in degrees):", round(angle_deg, 2))
```

```python
# 🎨 3. Visualize Vectors in 2D using Matplotlib
import matplotlib.pyplot as plt

origin = np.zeros(2)

plt.quiver(*origin, a[0], a[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector a')
plt.quiver(*origin, b[0], b[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector b')

plt.xlim(0, 6)
plt.ylim(0, 6)
plt.grid(True)
plt.legend()
plt.title("2D Vector Visualization")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

### 🧠 Reflection
- Notice how both vectors share similar directions (small angle).  
- Visualizing helps you *see* what dot product and magnitude truly mean — direction, similarity, and scale.


## 🧩 Challenge Task — Vector Relationship Analyzer
Now that you understand vector direction and magnitude, let's write a function that can interpret their relationship automatically.


```python
# ⚙️ Vector Relationship Function

import numpy as np

def compare_vectors(a, b):
    dot = np.dot(a, b)
    norm_a = np.linalg.norm(a)
    norm_b = np.linalg.norm(b)
    cos_theta = dot / (norm_a * norm_b)
    
    # Numerical tolerance for floating-point comparison
    if np.isclose(cos_theta, 1, atol=1e-3):
        relation = "✅ Same Direction"
    elif np.isclose(cos_theta, -1, atol=1e-3):
        relation = "🔁 Opposite Direction"
    elif np.isclose(cos_theta, 0, atol=1e-3):
        relation = "📐 Perpendicular"
    else:
        relation = "🔄 Some Angle Between (Neither parallel nor perpendicular)"
    
    angle_deg = np.degrees(np.arccos(np.clip(cos_theta, -1.0, 1.0)))
    print(f"Dot Product: {dot}")
    print(f"Angle Between Vectors: {round(angle_deg, 2)}°")
    print(f"Relation: {relation}")

# Example tests
compare_vectors(np.array([1, 2]), np.array([2, 4]))
compare_vectors(np.array([1, 0]), np.array([-1, 0]))
compare_vectors(np.array([0, 1]), np.array([1, 0]))
```

### 🧠 Reflection
- The function uses **dot product** and **cosine similarity** to detect relationships between vectors.  
- This concept is the mathematical heart of **AI embeddings** — like how ChatGPT measures similarity between two sentences in vector space!


## 📈 Bonus Visualization — Parallel, Opposite, and Perpendicular Vectors
Let’s visualize all three vector relationships in a single plot.
This will help you “see” how dot product and cosine angle translate into geometry.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the three vector pairs
origin = np.zeros(2)
v_parallel_1 = np.array([2, 1])
v_parallel_2 = np.array([4, 2])

v_opposite_1 = np.array([2, 0])
v_opposite_2 = np.array([-2, 0])

v_perp_1 = np.array([2, 0])
v_perp_2 = np.array([0, 2])

# Create a subplot layout
fig, axes = plt.subplots(1, 3, figsize=(14, 4))
titles = ["Parallel Vectors", "Opposite Vectors", "Perpendicular Vectors"]
colors = [("r", "g"), ("b", "orange"), ("purple", "teal")]

pairs = [
    (v_parallel_1, v_parallel_2),
    (v_opposite_1, v_opposite_2),
    (v_perp_1, v_perp_2)
]

for ax, (v1, v2), title, (c1, c2) in zip(axes, pairs, titles, colors):
    ax.quiver(*origin, *v1, angles='xy', scale_units='xy', scale=1, color=c1, label='v1')
    ax.quiver(*origin, *v2, angles='xy', scale_units='xy', scale=1, color=c2, label='v2')
    ax.set_xlim(-3, 5)
    ax.set_ylim(-3, 5)
    ax.grid(True)
    ax.legend()
    ax.set_title(title)
    ax.set_xlabel("X-axis")
    ax.set_ylabel("Y-axis")

plt.suptitle("Visualizing Vector Relationships", fontsize=14, fontweight="bold")
plt.show()
```

### 🔍 Insights
- **Parallel vectors** point in the same direction — dot product is **positive & large**  
- **Opposite vectors** point in opposite directions — dot product is **negative**  
- **Perpendicular vectors** are at 90° — dot product is **zero**
